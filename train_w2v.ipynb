```python
import gensim.models.keyedvectors as word2vec
w2v_model = word2vec.KeyedVectors.load('/content/drive/MyDrive/text summarization /vietnamese-text-summarization-master/sources/model/w2v_sen1.model')

vocabulary = []
for word in w2v_model.vocab.keys():
    vocabulary.append(word)
print(len(vocabulary))
```

```python
print(vocabulary[4])
```

```python
keys = ['đẹp', 'xấu']

embedding_clusters = []
word_clusters = []
for word in keys:
    embeddings = []
    words = []
    for similar_word, _ in w2v_model.most_similar(word, topn=30):
        words.append(similar_word)
        embeddings.append(w2v_model[similar_word])
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
```

```python
for word in keys:
    embeddings = []
    words = []
    embeddings.append(w2v_model[word])
    words.append(word)
    
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
    
print(word_clusters)
```
